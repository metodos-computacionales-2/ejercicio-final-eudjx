```python
import numpy as np
import matplotlib.pylab as plt
from IPython.display import clear_output
data= np.genfromtxt("data.dat")
t= data[:,0]
th= data[:,1]
w= data[:,2]
marker= data[:,3]
```

```python
indice=[]
for i in range(len(marker)):
    if(marker[i]==1):
        indice.append(i+1)
for i in range(len(indice)):
    clear_output(wait=True)
    if(i==0):
        Fd=1.4
        plt.figure(figsize=(15,15))
        plt.plot(th[:indice[i]],w[:indice[i]],".",label='F= '+str(round(Fd,2)))
        Fd+=0.01
    if(i>0):
        plt.figure(figsize=(15,15))
        plt.plot(th[indice[i-1]:indice[i]],w[indice[i-1]:indice[i]],".",label='F= '+str(round(Fd,2)))
        Fd+=0.01
    plt.legend()
    plt.savefig(str(i).zfill(4)+".png")
    plt.show()
    plt.close()
```
