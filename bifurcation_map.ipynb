```python
import numpy as np
import matplotlib.pylab as plt
from IPython.display import clear_output
data= np.genfromtxt("data.dat")
th= data[:,0]
Fd= data[:,1]
```

```python
plt.figure(figsize=(15,15))
plt.plot(Fd,th,".",ms=3)
plt.xlim(1.4275,1.50)
plt.grid()
plt.ylabel('$\Theta$($F_e$)',size=15)
plt.xlabel('$F_e$',size=20)
plt.title("Bifurcation map for Pendulum model",size=25)
plt.savefig("Bifurcation map.png")
plt.show()
```
