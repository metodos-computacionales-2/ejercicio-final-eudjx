```python
import numpy as np
import matplotlib.pylab as plt
from IPython.display import clear_output
data= np.genfromtxt("data.dat")
t= data[:,0]
th= data[:,1]
w= data[:,2]
N=int(len(data)/10000)
```

```python
N_T= 10000
for i in range(1,N+1):
    clear_output(wait=True)
    if(i==1):
        Fd=1.4
        plt.figure(figsize=(15,15))
        plt.plot(t[:N_T],th[:N_T],label='F= '+str(round(Fd,2)))
        Fd+=0.01
    if(i>1):
        lw= (N_T*(i-1))
        lu= (N_T*(i))
        plt.figure(figsize=(15,15))
        plt.plot(t[lw:lu],th[lw:lu],label='F= '+str(round(Fd,2)))
        Fd+=0.01
    plt.legend()
    plt.savefig(str(i).zfill(4)+".png")
    plt.show()
```
